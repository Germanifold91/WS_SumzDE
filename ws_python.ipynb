# 5. Preguntas

1. Quisiera saber cuáles fueron los productos más populares cada mes durante el año 2018. ¿Me puedes mostrar esa información?

```python
#Conversion de columna tipo 'object' a tipo datetime y extraccion de año y mes de cada fecha respectiva
sales_df['date'] = pd.to_datetime(sales_df['date'])
sales_df['year'] = sales_df['date'].dt.year
sales_df['month'] = sales_df['date'].dt.month
```

```python
sales_2018 = sales_df[sales_df['year'] == 2018].groupby(['month', 'product_id']).agg(volumen = ('sales', 'sum'),
                                                                     ingresos = ('revenue', 'sum')).reset_index()
sales_2018 = sales_2018.groupby('month', group_keys=False).apply(lambda x: x.sort_values('volumen', ascending=False))
```

```python
#RESPUESTA
#Top 5 de productos por mes mas populares por volumen para el año 2018
sales_2018.groupby('month').head(5).head(20)
```

2. Necesito saber en cuántas tiendas se implementaron promociones del tipo 1 en el canal 1 durante el año 2019. ¿Puedes darme ese dato?</li>

```python
#RESPUESTA
sales_df[(sales_df['year'] == 2019) & (sales_df['promo_type_1'].notnull()) & (sales_df['promo_bin_1'].notnull())]['store_id'].nunique()
```

3. Me gustaría saber cuál fue la ciudad que tuvo el mayor volumen de ventas en general durante todo el periodo de 2017 a 2019. ¿Tienes esa información disponible?

```python
store_sales = sales_df[sales_df['year'].isin([2017, 2018, 2019])].groupby('store_id')['sales'].sum().reset_index()
store_sales = pd.merge(store_sales, stores_df[['store_id', 'city_id']], on = 'store_id', how = 'left')
store_sales.groupby('city_id')['sales'].sum().reset_index().sort_values('sales', ascending=False).head(10)
```

4. ¿Podrías decirme cuál fue el promedio de stock disponible en todas las tiendas durante el año 2017? Estoy interesado en conocer ese dato.

6. ¿Cuál fue el producto con el precio de venta más alto en cada tienda durante el tercer trimestre de 2019? Me gustaría conocer esos detalles.

```python
price_hist = sales_df[(sales_df['date'] >= '2019-07-01') & (sales_df['date'] <= '2019-09-30')].groupby(['store_id', 'product_id'])['price'].max().reset_index()
price_hist = price_hist.groupby('store_id', group_keys=False).apply(lambda x: x.sort_values('price', ascending=False))
```
