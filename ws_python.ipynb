# 5. Preguntas

1. Quisiera saber cuáles fueron los productos más populares cada mes durante el año 2018. ¿Me puedes mostrar esa información?

```python
#Conversion de columna tipo 'object' a tipo datetime y extraccion de año y mes de cada fecha respectiva
sales_df['date'] = pd.to_datetime(sales_df['date'])
sales_df['year'] = sales_df['date'].dt.year
sales_df['month'] = sales_df['date'].dt.month
```

```python
sales_df
```

```python
sales_2018 = sales_df[sales_df['year'] == 2018].groupby(['month', 'product_id']).agg(volumen = ('sales', 'sum'),
                                                                     ingresos = ('revenue', 'sum')).reset_index()
sales_2018 = sales_2018.groupby('month', group_keys=False).apply(lambda x: x.sort_values('volumen', ascending=False))
```

```python
#RESPUESTA
#Top 5 de productos por mes mas populares por volumen para el año 2018
sales_2018.groupby('month').head(5).head(20)
```

3. Me gustaría saber cuál fue la ciudad que tuvo el mayor volumen de ventas en general durante todo el periodo de 2017 a 2019. ¿Tienes esa información disponible?

6. ¿Cuál fue el producto con el precio de venta más alto en cada tienda durante el tercer trimestre de 2019? Me gustaría conocer esos detalles.

```python
price_hist = sales_df[(sales_df['date'] >= '2019-07-01') & (sales_df['date'] <= '2019-09-30')].groupby(['store_id', 'product_id'])['price'].max().reset_index()
price_hist = price_hist.groupby('store_id', group_keys=False).apply(lambda x: x.sort_values('price', ascending=False))
```

```python
#RESPUESTA
price_hist.groupby('store_id').head(1).sort_values('store_id').head(20)
```

7. Estoy interesado en saber cuántas tiendas experimentaron un aumento de ventas del 20% o más en comparación con el año anterior. ¿Tienes esos datos disponibles?

```python
sales_hist = sales_df.groupby(['store_id', 'year'])['revenue'].sum().reset_index().sort_values(['store_id', 'year'])
sales_hist['previous_year_sales'] = sales_hist.groupby(['store_id'])['revenue'].shift(1)
sales_hist['change_previous_year'] = (sales_hist['revenue'] - sales_hist['previous_year_sales'])/sales_hist['previous_year_sales']
sales_hist = sales_hist[sales_hist['change_previous_year'].notnull()]
```

```python
sales_hist.head(10)
```

```python
#RESPUESTA
v1 = sales_hist[(sales_hist['year'] == 2019) & 
                (sales_hist['change_previous_year'] >= 0.2)]['store_id'].nunique()
v2 = sales_hist[sales_hist['change_previous_year'] >= 0.2]['store_id'].nunique()
print(f"Numero de tiendas que aumentaron sus ventas entre el 2018 al 2019 en al menos un 20%: {v1} \nNumero de tiendas que aumentaron sus ventas en comparasion con el año anterior en al menos un 20%: {v2}")
```

8. Quisiera saber cuál fue el producto que experimentó la mayor variación de precio a lo largo del periodo de 2017 a 2019. ¿Puedes proporcionarme esa información?

```python
price_hist = sales_df.groupby(['product_id', 'year'])['price'].mean().reset_index()
price_hist['initial_price'] = price_hist.groupby(['product_id'])['price'].shift(2)
price_hist = price_hist[price_hist['initial_price'].notnull()]
price_hist['price_variation'] = (price_hist['price'] - price_hist['initial_price'])/price_hist['initial_price']
```

```python
#RESPUESTA
price_hist.sort_values('price_variation', ascending=False)
```
